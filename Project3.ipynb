Determine Range for Time window

```
#Determine range for years
#Create Start and Stop Index

#2020
#end20 = 0
start20 = T[T>='2020-01-01T00'].index.tolist()
start20 = np.max(start20)
#T[start20]
#2019
end19 =start20+1
start19 = T[T>='2019-01-01T00'].index.tolist()
start19 = np.max(start19)
#2018
end18 = start19+1
start18 = T[T>='2018-01-01T00'].index.tolist()
start18 = np.max(start18)
#2017
end17 = start18+1
start17 = T[T>='2017-01-01T00'].index.tolist()
start17 = np.max(start17)
#2016
end16 = start17+1
start16 = T[T>='2016-01-01T00'].index.tolist()
start16 = np.max(start16)
#2015
end15 = start16+1
start15 = T[T>='2015-01-01T00'].index.tolist()
start15 = np.max(start15)
#2014
end14 = start15+1
start14 = T[T>='2014-01-01T00'].index.tolist()
start14 = np.max(start14)
#2013
end13 = start14+1
start13 = T[T>='2013-01-01T00'].index.tolist()
start13 = np.max(start13)
#2012
end12 = start13+1
start12 = T[T>='2012-01-01T00'].index.tolist()
start12 = np.max(start12)
#2011
end11 = start12+1
start11 = T[T>='2011-01-01T00'].index.tolist()
start11 = np.max(start11)
#2010
end10 = start11+1
#start10 = T[T>='2017-01-01T00'].index.tolist()
start10 = len(T)
```

```

df = data
df.head()

df['text'] = df['time'] + '<br>Magnitude ' + (df['mag']).astype(str)
limits = [(0,start20),(end19,start19),(end18,start18),(end17,start17),(end16,start16),(end15,start15),(end14,start14),(end13,start13),(end12,start12),(end11,start11),(end10,start10)]
colors = ["royalblue","crimson","lightseagreen","orange","lightgrey","royalblue","crimson","lightseagreen","orange","lightgrey","royalblue"]
year = [(2020),(2019),(2018),(2017),(2016),(2015),(2014),(2013),(2012),(2011),(2010)]

fig = go.Figure()

for i in range(len(limits)):
    yr = year[i]

for i in range(len(limits)):
    lim = limits[i]
    
    df_sub = df[lim[0]:lim[1]]
    fig.add_trace(go.Scattergeo(
        locationmode = 'ISO-3',
        lon = df_sub['longitude'],
        lat = df_sub['latitude'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['mag'],
            color = colors[i],
            line_color='rgb(40,40,40)',
            line_width=0.5,
            sizemode = 'area'
        ),
     name = '{0}'.format(year[i])))

fig.update_layout(
        title_text = 'Seismic Data in The Pacific Northwest 2010-2020<br>(Click legend to toggle years on and off)',
        showlegend = True,
        
        geo = dict(
            scope = 'north america',
            landcolor = 'rgb(217, 217, 217)',
        )
    )

fig.show()
```

#Map earthquakes in April 2015

```
#Create Start and Stop Index
April_end = T[T<='2015-05-01T00'].index.tolist()
April_end = np.min(April_end)
#T[April_end]
April_start = T[T>='2015-04-01T00'].index.tolist()
April_start = np.max(April_start)
#T[April_start]
```

```
#Create Data Sets for Plot
T15 = T[April_end:April_start]
Lat15 = Lat[April_end:April_start]
Lon15 = Lon[April_end:April_start]
Mag15 = Mag[April_end:April_start]
Mag15_avg = sum(Mag15)/len(Mag15)
```

```
#Map earthquakes in April 2015
fig, ax = plt.subplots(figsize = (9,8))
ax.scatter(Lon15, Lat15, zorder=1, alpha= 0.5, c='r', s= Mag15*10/Mag15_avg, label = 'Seismic Activity')
ax.set_title('Earthquake Magnitudes April 2015')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.legend()
ax.imshow(all_map, zorder=0, extent = BBox, aspect= 'equal')
```
